### Importing Libraries

```python
import cv2
import matplotlib.pyplot as plt
```

###  Uploading Original Image

```python
image = cv2.imread("C://Users//hp//Downloads//cat2.jpg")
```

```python
image2 =cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image2)
plt.axis(False)
plt.show()
```

### Converting Original Image to Grey Scale

```python
grey_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.axis(False)
plt.imshow(grey_image)
```

### Inverting The Grey Scale Image

```python
Invert_image = cv2.bitwise_not(grey_image)
plt.axis(False)
plt.imshow(Invert_image)
```

### Inverting The Blurred Image

```python
blur_image = cv2.GaussianBlur(Invert_image, (111,111),0)
plt.axis(False)
plt.imshow(blur_image)
```

```python
Pencil_image = cv2.divide(grey_image, blur_image, scale=256.0)
plt.axis(False)
plt.imshow(Pencil_image)
```

```python
cv2.imwrite("pencil.png",Pencil_image )
```

### Pencil Sketch

```python
plt.figure(figsize = (8,8))
Final_image= cv2.cvtColor(Pencil_image, cv2.COLOR_RGB2BGR)
plt.imshow(Final_image)
plt.axis(False)
plt.show()
```
